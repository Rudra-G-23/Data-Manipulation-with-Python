## find ranking of each movie in the group according to IMDB score


```python
movies['Rating'].rank(ascending=False)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

## find normalized IMDB rating group wise

```python
def normalized_rating(gp):
   gp['norm_rating'] =  ( (gp['Rating']) - (gp['Rating'].min()) ) /  ( (gp['Rating'].max()) - (gp['Rating'].min()) )
   return gp

genres.apply(normalized_rating)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

## Create a function that can return the highest score of any batsman

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
ipl.columns
```

```python
ipl['batsman'].unique()
```

```python
def highest_score_of_batsman(batsman_name):
    df = ipl[ipl['batsman'] == batsman_name]   
    df = df.groupby('match_id')['batsman_runs'].sum()
    return df.nlargest(10)
```

```python
highest_score_of_batsman('Mohammed Shami')
```

```python
highest_score_of_batsman('S Dhawan')
```

```python
highest_score_of_batsman('V Kohli')
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```
